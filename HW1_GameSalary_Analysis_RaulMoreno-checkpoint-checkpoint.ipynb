```python
# Python Programming 

# List and function
# Finish a function to user enter quarterlty sales amount.
# Finish a function to display sales
# In main fuction, invoke above functions and shw total and average sales
# Show the result as the example

QUARTER =["Spring", "Summer", "Fall", "Winter"]
           
def main():
    # create a list to hold the sales of four quarters 
    sales =[0.0] * len(QUARTER)

    inputSales(sales)
    
    print('\nDisplay the quarterly sales')
    display (sales)
    
    print('\nDisplay the summary')
    print('Total sales:              $', sum(sales))
    print('Average quarterly sales   $', sum(sales)/len(QUARTER))

# Ask user to enter sales:
def inputSales(sales):
    for i in range(0, len(sales)):
        print('Enter the sales of', QUARTER[i] ,  ' : $' , end = ' ')
        sales[i] = float(input())

# 1) Finish the method here:
def display (sales):
    for i in range(0, len(sales)):
        print('Sales of', QUARTER[i] , ' : $' ,  sales[i] , end = ' \n')


        
# Invoke 
main()
```

```python
# 14)
# A) Show all Titles name with no duplicate,       
# B) Display the title names as flowing format    

#res = gameSalary['Title'].unique()
res = gameSalary.Title.unique()
print(res)

print('\nTitles')
print('----------')
for name in res:
    print(name)
```

```python
# 15) Create a bar chart to compare average Salary by gender. 
# .groupby('')      
# .mean()      
# plot    

res = gameSalary.groupby('Gender')['Salary'].mean()

res.plot(kind='bar')
plt.title('Average Salary by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Salary')
plt.show()
```

```python
# 16)+17) 
# Create a bar chart to compare average Salary by Department. 
# .groupby('')      
# .mean()      
# plot    


res = gameSalary.groupby('Department')['Salary'].mean()

res.plot(kind='bar')
plt.title('Average Salary by Department')
plt.xlabel('Department')
plt.ylabel('Average Salary')
plt.show()


```

```python
# 17) + 18) Show the number of employees for each performance level 
# and create pie chart (and show percentage)

res = gameSalary["Performance"].value_counts()

print(res)

res.plot(kind = "pie", autopct = '%1.1f%%' )

plt.title('Employees by Performance Level')
plt.show()


```

```python
# 19) + 20) 
# Show the number of employee for each department,
# and sorted by the values from hight to low
# And plot the result as bar chart

#  .groupby('Department') or df.Department.value_counts() 
#  .sort_values(ascending=False)  
#  .plot 

res = gameSalary["Department"].value_counts().sort_values(ascending=False)

print(res)

res.plot(kind='bar')
plt.title('Number of Employees by Department')
plt.xlabel('Department')
plt.ylabel('Number of Employees')
plt.show()
```
